## 3. Value-based Metrics Summary

```
action_counts = targeting_list['recommended_action'].value_counts()
print('Targeting action breakdown:\n', action_counts)

targeted = targeting_list[targeting_list['recommended_action'] == 'TARGET']
total_incremental_rev = targeted['expected_incremental_revenue'].sum()
total_incremental_profit = targeted['expected_incremental_profit'].sum()

print(f'\nProjected Business Outcome:')
print(f'- Total targeted customers: {len(targeted)}')
print(f'- Expected Incremental Revenue: {total_incremental_rev:,.2f} THB')
print(f'- Expected Incremental Net Profit: {total_incremental_profit:,.2f} THB')
```
