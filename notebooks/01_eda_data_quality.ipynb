# 📊 Part 1: Exploratory Data Analysis & Data Quality Check
This notebook performs initial EDA and data quality checks on the mock retail datasets.

**Goal:** Understand customer segments, transaction distributions, and ensure no missing values or duplicates exist before modeling.

```
import sys
import os
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_loader import load_data
sns.set_theme(style='minimal')
```

## 1. Load Data

```
dfs = load_data(data_dir='../data')
for name, df in dfs.items():
    print(f"{name.upper()}: {df.shape[0]} rows, {df.shape[1]} columns")
```

## 2. Check Data Quality
Check for missing values, duplicates, and general stats.

```
for name, df in dfs.items():
    print(f"--- {name.upper()} --- ")
    print("Missing Values:\n", df.isnull().sum())
    print("Duplicate Rows:", df.duplicated().sum())
    print()
```

## 3. Customer Segments Distribution

```
plt.figure(figsize=(8, 4))
sns.countplot(data=dfs['customers'], x='customer_taxonomies', palette='viridis')
plt.title('Customer Segment Distribution')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.show()
```

## 4. Sales Transaction Trend

```
tx = dfs['transactions'].copy()
tx['datetime'] = pd.to_datetime(tx['datetime'])
tx['date'] = tx['datetime'].dt.date
daily_tx = tx.groupby('date')['po_id'].nunique().reset_index()

plt.figure(figsize=(12, 4))
plt.plot(daily_tx['date'], daily_tx['po_id'], color='coral')
plt.title('Daily Unique Transactions')
plt.xlabel('Date')
plt.ylabel('Transactions')
plt.grid(True, linestyle='--')
plt.show()
```
